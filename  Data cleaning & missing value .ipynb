```python
## messy data
employees=[("stuti",28,"F",""),
           ("saumya",32,"F",""),
           ("Aaditya",25,"","m"),
           ("saumya",32,"Fe",""),
           ("saumya",32,"Fe",""),
           ("saumya",32,"F",""),
           ("Aaditya",40,"","m"),
           ("seema",32,"F","")
            ]
df=pd.DataFrame(employees,columns=["name","age","isfemale","ismale"])
df
```
