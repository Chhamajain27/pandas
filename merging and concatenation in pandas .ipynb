# Merging

```python
# merging is used when atleast one coulmn is same and it is merged column wise and  
```

```python
import pandas as pd
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
```

```python
df1
```

```python
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})


df2
```

```python
pd.merge(df1,df2,how="inner",on="key")
```

```python
pd.merge(df1,df2,how="left",on="key")
```

```python
pd.merge(df1,df2,how="right",on="key")
```

```python
pd.merge(df1,df2,how="outer",on="key")
```

```python
# when rows are different in both the dataframe
import pandas as pd
df11 = pd.DataFrame({'key': ['K0', 'K8', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
   
df22 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'c': ['C0', 'C1', 'C2', 'C3'],
                          'd': ['D0', 'D1', 'D2', 'D3']})

```

```python
df11
```

```python
df22
```

```python
pd.merge(df11,df22,how="inner",on="key")
```

```python
pd.merge(df11,df22,how="left",on="key")
```

```python
pd.merge(df11,df22,how="right",on="key")
```

```python
# merging on more days one key: key1 and key2 both as below
df1 = pd.DataFrame({'key1': ['K0', 'K8', 'K2', 'K3'],
                    'key2': ['K10', 'K18', 'K12', 'K13'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                    'key2': ['K10', 'K11', 'K12', 'K13'],
                          'A': ['C0', 'C1', 'C2', 'C3'],
                          'B': ['D0', 'D1', 'D2', 'D3']})

```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how="left",on=["key1","key2"])
```

```python
pd.merge(df1,df2,how="inner",on=["key1","key2"])
```

```python
pd.merge(df1,df2,how="right",on=["key1","key2"])
```

```python
# concat combining two or more dataframes horizontally or vertically
### 
```

```python
## taking three data frames df1 ,df2 and df3
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
print(df1,"\n\n")
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
print(df2,"\n\n")
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8,9,10,11])
print(df3,"\n\n")


```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```
